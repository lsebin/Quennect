```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('q_raw.csv')
df.info()
```

```python
df.head(10)
```

```python
encoded_df = pd.read_csv('q_data_encoded.csv')
encoded_df.head(10)
```

```python
columns_to_keep = ['q_date', 'q_year', 'prop_year', 'ia_status_raw', 'region','state', 'type1','mw1', 'mw2', 'mw3']
df_dropped = df[columns_to_keep].copy()
```

```python
# categorical = ["state", "region", "type1"]
value_counts_region = df_dropped['region'].value_counts().head(10)
print(value_counts_region)
```

```python
# One-hot encode the new column
encoded_categories = pd.get_dummies(df_dropped['region'], prefix='region').astype(int)

# Concatenate the one-hot encoded columns with the original DataFrame
df_with_region = pd.concat([df_dropped, encoded_categories], axis=1)
df_with_region.head(10)
```

```python
df_with_region['ia_status_category'] = df_with_region['ia_status_raw'].replace({
    'System Impact Study': 'System Impact Study',
    'IA Executed': 'IA Executed',
    'Executed': 'IA Executed',
    'Withdrawn': 'Withdrawn',
    'Feasibility Study': 'Feasibility Study',
    'Feasibility': 'Feasibility Study',
    'In progress': 'In progress',
    'Active': 'Operational',
    'Operational': 'Operational',
    'operational': 'Operational',
    'Facility Study': 'Facility Study'
})

df_revised = df_with_region[df_with_region['ia_status_category'].isin([
    'System Impact Study',
    'IA Executed',
    'Withdrawn',
    'Feasibility Study',
    'In progress',
    'Operational',
    'Facility Study'
])]

df_revised.head(10)
```

```python
columns_to_sum = ['mw1', 'mw2', 'mw3']
df_revised['total_mw'] = df_revised[columns_to_sum].sum(axis=1, skipna=True)
df_cleaner = df_revised.copy().drop(columns=['mw1', 'mw2','mw3']).dropna()
df_cleaner.head(10)
```

```python
types = pd.get_dummies(df_cleaner['type1'], prefix='type1').astype(int)

# Concatenate the one-hot encoded columns with the original DataFrame
df_cleaned = pd.concat([df_cleaner,types], axis=1)
df_cleaned.head(10)
```

```python
types = pd.get_dummies(df_cleaned['ia_status_category'], prefix='ia_status').astype(int)

# Concatenate the one-hot encoded columns with the original DataFrame
df_final_dropped= pd.concat([df_cleaned,types], axis=1)
df_final_dropped.head(10)
```

```python
df_final_dropped = df_final_dropped.drop(columns=['q_date', 'q_year',	'prop_year', 'ia_status_raw','ia_status_category', 'region', 'state', 'type1','ia_status_Facility Study',
       'ia_status_Feasibility Study', 'ia_status_IA Executed',
       'ia_status_Operational', 'ia_status_System Impact Study'])
# df_final_dropped = df_cleaned.drop(columns=['q_date', 'q_year',	'prop_year', 'ia_status_raw', 'region', 'state', 'type1'])
# df_final_dropped.head(10)
```

```python
df_final_dropped = df_final_dropped.drop(columns=['type1_Unknown', 'type1_Other', 'type1_Other Storage'])
```
